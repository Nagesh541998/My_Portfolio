## Data Visualization

Data visualization is a graphical representation of information of data. By using visual elements like charts and graphs, Data visualization libraries provide an accessible way to see and understand trends, outliers, and patterns in data.

### Matplotlib

### Load Necessary Libraries

```python
import matplotlib.pyplot as plt     # Visualization
import pandas as pd                 # Data Manipulation
import numpy as np                  # Numerical python for scientific computing
```

### Basic Graph

#### Line Graph is commonly used to show trends or changes in data over time.

Syntax for Line Graph

------  plt.plot(x,y)      by default plot takes line plot, Both x & y Should be 1D array

>>> plot(x, y, 'go--', linewidth=2, markersize=12)
>>> plot(x, y, color='green', marker='o', linestyle='dashed',
...      linewidth=2, markersize=12)

```python
x = [2,4,6,8,10,12]
y = [1,2,3,4,5,6]

plt.plot(x,y,'go--',linewidth = 2,markersize = 12)
plt.show()
```

```python
# Now will plot the same graph by giving some title,labels,ticks etc

plt.figure(figsize=(6,3),dpi=100)     #dpi stands for dots per inches it gives the pixels of figure the default dpi is 100

plt.plot([1,2,3],[2,4,6],label='2x',color='blue',linestyle='-',linewidth=2,marker='*',markersize=10,markeredgecolor='red')


# will perform 2nd plot
x=np.arange(1,3.5,0.5)
plt.plot(x[:3],x[:3]**2,color = 'r',marker='^')
plt.plot(x[2:],x[2:]**2,'r--')

plt.title('Line Plot',fontdict={'fontname':'Times New Roman'})  #we can give title and even the font name and size
plt.xlabel('Predictors',fontdict = {'fontname':'Times New Roman'}) 
plt.ylabel('Target',fontdict = {'fontname':'Times New Roman'})
plt.xticks(list(range(1,4)))
plt.yticks(list(range(2,7,2)))
plt.legend()
plt.show()
```

### Pie Chart
#### Pie chart is a graphical representation of the data it shows how a total amount is divided between levels of categorical variable as a circle divided 

```python
exp_vals=[1400,600,410,250]
exp_lables=["house rent","food","shares","other"]
plt.pie(exp_vals,labels=exp_lables,autopct="%.2f %%")

plt.title("Expenses",fontdict={"fontname":"Times new roman","fontsize":30,"fontweight":"bold"})
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
