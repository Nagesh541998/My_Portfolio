### DataFrame Sorting

```python
# Head is used to print top 5 records (tail is used to print bottom 5 records)
import pandas as pd
df1 = pd.read_csv('bigmarket.csv')
df1.head()
```

```python
# It sort the data in ascending order of sales and it doesn't affect the original data
df1.sort_values('Sales')
```

```python
# It sort the data in descending value of sales and it doesn't affect the original data
df1.sort_values('Sales',ascending=False)
```

```python
df=df1.copy()
```

```python
# Sorting the data based on Store desc
df.sort_values('Store',ascending = False,inplace=True)
```

```python
# Now will reset the index (it create new index column insted or replacing it)
df.reset_index()
```

```python
# If you want to replace and affect the original index
df.reset_index(drop=True,inplace=True)
```

```python
df.head()    # If you want to affect the original data (inplace=True)
```

#### If you want to sort based on 2 columns

```python
df1.sort_values(['Store','Sales'],ascending=[False,True]).reset_index(drop=True)
```

### Concatenating DataFrame (appending the dataframe)

```python
# If you use axis=0 it acts as appending (it concat row wise)
new_df=pd.concat([cust_data,ord_data],axis=0,ignore_index=True)
new_df
```

```python
# If you use axis=1 (it concat column wise)
new_df = pd.concat([cust_data,ord_data],axis=1,ignore_index=True)
new_df
```

####  Inner Join

```python
new_df_inner = pd.merge(cust_data,ord_data,on='Cust_ID',how='inner')
new_df_inner
```

#### Left Join

```python
new_df_left = pd.merge(cust_data,ord_data,on='Cust_ID',how='left')
new_df_left
```

#### Right Join

```python
new_df_right = pd.merge(cust_data,ord_data,on='Cust_ID',how='right')
new_df_right
```

```python
# If i have both joining column with different name 
# Then we use left_on and right_on
import pandas as pd
inner = pd.merge(cust_data,ord_data,left_on='Cust_ID',right_on='Cust_ID',
                how='inner',suffixes=('_cust','_ord'))
inner
```
